```python
past_officers.head()
# Get column of list of winners 
winners = past_officers["Name"]
```

```python
def list_contains(List1, List2):
    isWinner = [] 
    # Iterate in the 1st list 
    for m in List1: 
        match = False
        for n in List2: 
            # if there is a match
            if m == n:
                match = True 
                break; 
        if match == True:
            isWinner.append(1)
        else:
            isWinner.append(0)          
    return isWinner
sanctions["isWinner"] = list_contains(sanctions["Candidate Name"], winners)
sanctions.head()
```

```python
winner_sanctions = sanctions[sanctions["isWinner"] == 1]
print(winner_sanctions["Candidate Name"].value_counts())
winner_sanctions[winner_sanctions["Year"] == 2020]
#winner_sanctions[winner_sanctions["Year"] == 2022]
```
